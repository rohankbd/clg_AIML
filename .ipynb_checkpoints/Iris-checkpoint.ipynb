```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
iris = datasets.load_iris()
print(iris)
```

```python
x = iris.data[:,:2] #we only take first 2 features
#we could avoid this ugly slicing by using a 2-D dataset
y = iris.target
print(x,y)
```

```python
#c -> SVM regularization parameter
svc = svm.SVC(kernel='linear',C=1).fit(x,y)
#create a mesh to plot in
x_min = x[:,0].min()-1
x_max = x[:,0].max()+1
y_min = x[:,1].min()-1
y_max = x[:,1].max()+1
print("xmin",x_min)
print("xmax",x_max)
print("ymin",y_min)
print("ymax",y_max)

h = (x_max/x_min)/100
print("h",h)
```

```python
xx, yy = np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
print("xx",xx)
print(xx.shape)
print(yy.shape)
print("yy",yy)
```

```python
z = svc.predict(np.c_[xx.ravel(),yy.ravel()])
z = z.reshape(xx.shape)
print("z",z)
```

```python
plt.contourf(xx,yy,z,cmap=plt.cm.Paired,alpha=0.4) #makes divisions(boundaries)

plt.scatter(x[:,0],x[:,1],c=y,cmap=plt.cm.Paired) #points scattered with different colors for data points

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('SVC with Linear Kernel')
plt.show()
```
